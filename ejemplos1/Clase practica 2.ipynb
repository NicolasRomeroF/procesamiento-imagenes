## Analisis basico
### Ejemplo 1

```python
#Tamaño de plot

samplingFrecuency=30 #cuantos puntos necesito
samplingInterval=1/samplingFrecuency #puntos muestreados
beginTime=0
endTime=10
signal1Frecuencia=4
signal2Frecuencia=7
time=np.arange(beginTime,endTime, samplingInterval)

amplitude1=np.sin(2*np.pi*signal1Frecuencia*time)
amplitude2=np.sin(2*np.pi*signal2Frecuencia*time)

#dibujar

fig,axis=plt.subplots(4,1)
plt.subplots_adjust(hspace=1)

axis[0].set_title("Senal seno de frecuencia 4")
axis[0].plot(time, amplitude1)
axis[0].set_xlabel("tiempo")
axis[0].set_ylabel("Amplitud")

axis[1].set_title("Senal seno de frecuencia 7")
axis[1].plot(time, amplitude2)
axis[1].set_xlabel("tiempo")
axis[1].set_ylabel("Amplitud")

#suma de señales
amplitude=amplitude1+amplitude2

axis[2].set_title("Suma")
axis[2].plot(time, amplitude)
axis[2].set_xlabel("tiempo")
axis[2].set_ylabel("Amplitud")

#señal en frecuencia

ft=np.fft.fft(amplitude)/len(amplitude)
ft=ft[range(int(len(amplitude)/2))]

tpCount=len(amplitude)
values=np.arange(int(tpCount/2))
timePeriod=tpCount/samplingFrecuency
frecuencies=values/timePeriod


axis[3].set_title("FFT")
axis[3].plot(frecuencies, abs(ft))
axis[3].set_xlabel("tiempo")
axis[3].set_ylabel("Amplitud")

plt.rcParams["figure.figsize"] = (10,10)
plt.show()
```

```python
f=5
f_s=100
t=np.linspace(0,1,1*f_s,endpoint=False)
x=np.sin(f*2*np.pi*t)

fig,ax=plt.subplots()
ax.plot(t,x)
ax.set_xlabel('Tiempo')
ax.set_ylabel('Amplitud')
```

```python
from scipy import fftpack

X=fftpack.fft(x)
freqs=fftpack.fftfreq(len(x))*f_s

fig,ax=plt.subplots()

ax.stem(freqs,np.abs(X))
ax.set_xlabel('Frecuencia en Hertz')
ax.set_ylabel('Spectrum - Magnitud')
ax.set_xlim(-f_s/2, f_s/2)
ax.set_ylim(-5,60)
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img=cv2.imread('vidal1.jpeg',0)
f=np.fft.fft2(img)
fshift=np.fft.fftshift(f)
magnitude_spectrum=20*np.log(np.abs(fshift))

plt.subplot(121),plt.imshow(img,cmap='gray')
plt.title('input'),plt.xticks([]),plt.yticks([])
plt.subplot(122), plt.imshow(magnitude_spectrum, cmap='gray')
plt.show()
```
