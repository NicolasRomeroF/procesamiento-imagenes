```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#cargar imagen o leer imagen
image = cv2.imread('vidal1.jpeg')
plt.imshow(image)
plt.title('Ventana vidal')
plt.show()
cv2.imwrite('vidal_azul.jpg', image)
```

## Propiedades de la imagen

```python
print('Propiedades de la imagen')
print(' Numero de pixeles: ' + str(image.size))
print(' Dimension: ' + str(image.shape))
```

## Convertir BGR a RGB

```python
image2 = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
plt.imshow(image2)
plt.title('Vidal normal')
plt.show()
cv2.imwrite('vidal_azul.jpg', image2)
```

```python
print('Propiedades de la imagen')
print(' Numero de pixeles: ' + str(image2.size))
print(' Dimension: ' + str(image2.shape))
```

## Concatenar imagenes

```python
con = np.concatenate((image,image2),axis=1)
plt.imshow(con)
plt.title('Concatenate vidal')
plt.show()
#axis 1= Horizontal, axis 0 = Vertical
```

```python
print('Propiedades de la imagen')
print(' Numero de pixeles: ' + str(con.size))
print(' Dimension: ' + str(con.shape))
```

```python
#Redimensionar la imagen manteniendo la relacion de aspecto
r = 100.0/image2.shape[1]
dim = (100, int(image2.shape[0]*r))
resized = cv2.resize(image2, dim, interpolation=cv2.INTER_AREA)
plt.imshow(resized)
plt.title('Vidal Resized')
plt.show()
```

```python
print('Propiedades de la imagen')
print(' Numero de pixeles: ' + str(resized.size))
print(' Dimension: ' + str(resized.shape))
```

```python
(h,w)= image2.shape[:2]
center = (w/2, h/2)
#rotar 180 grados
M = cv2.getRotationMatrix2D(center, 180, 1.0)
rotated = cv2.warpAffine(image2, M, (w,h))
plt.imshow(rotated)
plt.title('Vidal de cabeza')
plt.show()
```

```python
new_vidal = cv2.flip(image2,-1)
plt.imshow(new_vidal)
```

## Mostra imagen por canales

```python
image = cv2.imread('vidal1.jpeg')
blue, green, red = cv2.split(image)
img_gs = cv2.imread('vidal1.jpeg', cv2.IMREAD_GRAYSCALE)

plt.imshow(image2)
plt.title('Vidal Normal')
plt.show()

plt.imshow(img_gs, cmap="gray")
plt.title('Vidal Gris')
plt.show()

plt.imshow(image2[:,:,0], cmap="Reds")
plt.title('Vidal red')
plt.show()

plt.imshow(image2[:,:,1], cmap="Greens")
plt.title('Vidal green')
plt.show()

plt.imshow(image2[:,:,2], cmap="Blues")
plt.title('Vidal blue')
plt.show()
```

```python
plt.subplot(221),plt.imshow(image2,'gray'), plt.title('Original')
plt.subplot(222),plt.imshow(red,'gray'), plt.title('Red')
plt.subplot(223),plt.imshow(green,'gray'), plt.title('Green')
plt.subplot(224),plt.imshow(blue,'gray'), plt.title('Blue')
```

## Definir una imagen vacia

```python
blank_img = np.zeros(shape=(512,512,3), dtype=np.int16)
plt.imshow(blank_img)
```

```python
cv2.circle(img = blank_img, center=(400,400), radius=50, color=(255,0,0), thickness=-1)
plt.imshow(blank_img)
```

```python
cv2.rectangle(img=blank_img, pt1=(384,0), pt2=(500,128), color=(255,0,0), thickness=100)
plt.imshow(blank_img)
```
